# Visualizing Time Series Data

```python
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib.dates as dates
%matplotlib inline
```

```python
df_apple = pd.read_csv('tsa_data/apple_stock.csv',index_col='Date',parse_dates=True)
print(df_apple.shape)
df_apple.head()
```

```python
df_apple[['Volume','Adj Close']].plot() # volumne    
df_apple[['Volume','Adj Close']].plot(secondary_y=['Volume'])
```

```python
df_apple['Adj Close'].plot(figsize=(12,8))
plt.ylabel('Close Price')
plt.xlabel('Overwrite Date')
plt.title('Apple')
```

# Plot Formatting
## X Limits

```python
df_apple['Adj Close'].plot(xlim=['2015-01-01','2018-01-01'])
```

## Y Limits

```python
df_apple['Adj Close'].plot(xlim=['2015-01-01','2018-01-01'],ylim=[80,180])
```

## Color and Style

```python
df_apple['Adj Close'].plot(xlim=['2015-01-01','2018-01-01'],ylim=[80,180],
                         ls='--',c='r') # linestyle / color
```

 ## Basic matplotlib plot

```python
idx = df_apple.loc['2015-01-01':'2018-01-01'].index
stock = df_apple.loc['2015-01-01':'2018-01-01']['Adj Close']

fig, ax = plt.subplots()
ax.plot_date(idx, stock,'-')
plt.tight_layout() #    
fig.autofmt_xdate() # label  
ax.yaxis.grid(True) # 
ax.xaxis.grid(True) # 

plt.show()
```

## Customize Format dates on Major Axis

```python
idx = df_apple.loc['2015-01-01':'2018-01-01'].index
stock = df_apple.loc['2015-01-01':'2018-01-01']['Adj Close']

fig, ax = plt.subplots(figsize=(20,10))
ax.plot_date(idx, stock,'-')
plt.tight_layout() #    
fig.autofmt_xdate() # label  
ax.yaxis.grid(True) # 
ax.xaxis.grid(True) # 

ax.xaxis.set_major_locator(dates.MonthLocator()) # X  FORAT
ax.xaxis.set_major_formatter(dates.DateFormatter('%b\n%Y'))

plt.show()

# Major Axis
```
